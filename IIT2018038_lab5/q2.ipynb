```python
import pandas as pd

df = pd.read_csv("q2_dataset.csv", index_col = 'RID')
print(df)

query = {"Age":"<=30" ,"Income":"Medium", "Student":"Yes" ,"Credit_rating":"Fair"}

print()
print("query = ", query)
print()
di = {}
li = list(df)
for i in li:
    di[i] = {}
    
print("li = ",li)
print()
output = 'Class: buys_computer'

for i in li:
    kk=0
    for j in df.index:
        if df[i][j] not in di[i]:
            di[i][df[i][j]] = kk
            if i!=output:
                kk+=1
                
print("di = ",di)
print()
dis = {}
for i in df.index:
    dd = 0
    for j in li:
        if j!=output:
            dd+=(di[j][df[j][i]]-di[j][query[j]])**2
    dis[i]=dd
print("distance-unsorted :", dis)
print()



sorted_dict = {}
sorted_keys = sorted(dis, key=dis.get)


for w in sorted_keys:
    sorted_dict[w] = dis[w]

print("distance-sorted :",sorted_dict) 

print()

k_val = 1
for i in sorted_dict:
    di[output][df[output][i]]+=1
    
    if di[output]['Yes']>di[output]['No']:
        print("For k=%d , query belongs to Class: buys_computer = "%k_val,'Yes')
    else:
        print("For k=%d , query belongs to Class: buys_computer = "%k_val,'No')
    k_val+=1
    
print()
print("Final di is", di)
```
