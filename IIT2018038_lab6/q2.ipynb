```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


def train_util(train, xmean, ymean):
   
    cov = 0
    var = 0
    for i in range(len(train)):
        cov+=(train["x"][i] - xmean)*(train["y"][i] - ymean)
        var+=(train["x"][i] - xmean)*(train["x"][i] - xmean)
        
    corel = cov/var
    return corel

########################################################################################
df = pd.read_csv("data - Sheet1.csv")
print(df.head())
print()

#let the best fit line be,  (y-ymean) = m*(x-xmean)


xmean = df["x"].mean()
ymean = df["y"].mean()
corel = train_util(df, xmean, ymean)
print("the best fit line is,  (y-%f) = %f*(x-%f)"%(ymean,corel,xmean))

####################################################################################################3
plt.scatter(df["x"],df["y"])
xx = np.linspace(0,10,100)
yy = corel*(xx-xmean) + ymean
plt.plot(xx,yy,color = 'r')
plt.show()
```
